```python
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
import tensorflow as tf
from tensorflow import keras
from keras.models import Sequential
from keras.layers import Dense, Flatten
```

```python
df=pd.read_csv("insurance_data.csv")
df.head()
```

```python
xtrain,xtest,ytrain,ytest=train_test_split(df[['age','affordibility']], df['bought_insurance'],test_size=0.2, random_state=25)
xtrain['age']=xtrain['age']/100
xtest['age']=xtest['age']/100
```

```python
class myNN():
    def __init__(self):
        
        pass
    
    def sigmoid(self,X):
        return (1/(1+np.exp(-X)))
        
    def gradient_dedcent(self,xtrain,ytrain, epochs, loss_treshold):
        w=np.ones(shape=xtrain.shape[1])
        n=xtrain.shape[0]
        bias=0
        rate=0.5        
        for i in range(epochs):
            weighted_sum = np.dot(w,xtrain.T)+bias
            y_predicted=self.sigmoid(weighted_sum)
            loss=np.mean(np.square(ytrain-y_predicted))

            wd=(-2/n)*np.dot((ytrain-y_predicted),xtrain)
            bias_d=(-2/n)*np.mean((ytrain-y_predicted))
            w =w-rate*wd
            bias=bias-rate*bias_d

            if i%50==0:
                print(f'Epoch:{i}, w:{w}, bias:{bias}, loss:{loss}')

            if loss <=loss_treshold:
                print(f'Epoch:{i}, w:{w}, bias:{bias}, loss:{loss}')
                break
        return w,bias

    
    def fit(self,xtrain,ytrain, epochs, loss_treshold):
        self.w, self.bias=self.gradient_dedcent(xtrain,ytrain, epochs, loss_treshold)

    
    def predict(self,xtest):
         weighted_sum =  np.dot(self.w,xtest.T)+self.bias
         return self.sigmoid(weighted_sum)
```

```python
mymodel=myNN()
mymodel.fit(xtrain,ytrain,1000,0.10)
```

```python
mymodel.predict(xtest)
```

```python
model=keras.Sequential([
    keras.layers.Dense(1,input_shape=(2,), activation='sigmoid'), 
]) 

model.compile(optimizer='adam',
              loss='binary_crossentropy',
               metrics=['accuracy'])

model.fit(xtrain, ytrain, epochs=5000)
```

```python
model.evaluate(xtest,ytest)
```
